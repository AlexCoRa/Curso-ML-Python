# Árboles de Regresión

```python
import pandas as pd
```

```python
data = pd.read_csv("../datasets/boston/Boston.csv")
data.head()
```

```python
data.shape
```

```python
colnames = data.columns.values.tolist()
predictors = colnames[:13]
target = colnames[13]
X = data[predictors]
Y = data[target]
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regtree = DecisionTreeRegressor(min_samples_split=30, max_depth=5 ,min_samples_leaf=10, random_state=0)
```

```python
regtree.fit(X,Y)
```

```python
preds = regtree.predict(data[predictors])
```

```python
data["preds"] = preds
```

```python
data[["preds", "medv"]]
```

```python
# Visualización del árbol de regresión
```

```python
from sklearn.tree import export_graphviz

with open("./resources/boston_rtree.dot", "w") as dotfile:
    export_graphviz(regtree, out_file=dotfile, feature_names=predictors)
    dotfile.close()

import os
from graphviz import Source

file= open("./resources/boston_rtree.dot")
text = file.read()

Source(text)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, make_scorer
import numpy as np
```

```python
cv = KFold(n_splits=10, shuffle=True, random_state=1)
scores = cross_val_score(regtree, X, Y, scoring=make_scorer(mean_squared_error), cv=cv, n_jobs=1)
print(scores)
score = np.mean(scores)
print(score)
```

```python
list(zip(predictors, regtree.feature_importances_))
```

# Random Forests

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor(n_jobs=2, oob_score=True, n_estimators=10000)
forest.fit(X,Y)
```

```python
data["rforest_pred"] = forest.oob_prediction_
data[["rforest_pred", "medv"]]
```

```python
data["rforest_error2"] = (data["rforest_pred"] - data["medv"])**2
sum(data["rforest_error2"]) / len(data)
```

# Resumen

* Los Árboles de decisión son un algoritmo de clasificacion que utiliza variables predictoras continuas o categoricas para predecir una categoria.
* Un nodo se divide en subnodos para obtener distribuciones homogeneas con tecnicas como la ganancia de informacion, el indice de Gini o la reduccion de la varianza.
* Los árboles de regresion son similares a los de decision pero en este caso la variable de decision es continua.
* Los bosques aleatorios son algoritmos de aprendizaje por combinacion donde combinamos diversos arboles para la decision final por mayoria o por promedio. Son mas eficientes que una sola decision ya que maximizan la reduccion de la varianza.
