# Agregación de datos por categoría

```python
import numpy as np
import pandas as pd
```

```python
genero = ["Masculino", "Femenino"]
ingresos = ["Baja", "Media", "Rica"]
```

```python
n = 500

genero_data = []
ingresos_data = []

for i in range(0,500):
    genero_data.append(np.random.choice(genero))
    ingresos_data.append(np.random.choice(ingresos))
```

```python
genero_data[:10]
```

```python
ingresos_data[:10]
```

* m -> media
* s -> desviacion tipica
* Z -> distribucion de numeros aleatorios de una normal 0,1

```python
#N(m, s) -> m + s * Z
estatura = 160 + 30 * np.random.randn(n)
peso = 65 + 20 * np.random.randn(n)
edad = (25 + 3 * np.random.randn(n)).astype(int)
ingresos = 18000 + 3500 * np.random.randn(n)
```

```python
data = pd.DataFrame(
    {
        "Genero": genero_data,
        "Estado Económico": ingresos_data,
        "Estatura (cm)": estatura,
        "Peso (kg)": peso,
        "Edad (años)": edad,
        "Ingresos ($)": ingresos
    }
)
```

```python
data.head()
```

## Agrupación de datos

```python
agrupar_genero = data.groupby("Genero")
```

```python
agrupar_genero.groups
```

```python
for nombres, grupos in agrupar_genero:
    print(nombres)
    print(grupos)
```

```python
agrupar_genero.get_group("Femenino")
```

```python
doble_agrupacion = data.groupby(["Genero", "Estado Económico"])
```

```python
len(doble_agrupacion)
```

```python
for nombres, grupos in doble_agrupacion:
    print(nombres)
    print(grupos)
```

```python
doble_agrupacion.aggregate(
    {
        "Ingresos ($)": np.sum,
        "Edad (años)": np.mean,
        "Peso (kg)": np.std
    }
)
```

```python
doble_agrupacion.aggregate(
    {
        "Edad (años)": np.mean,
        "Estatura (cm)": lambda a:(np.mean(a))/np.std(a)
    }
)
```

```python
doble_agrupacion.aggregate([np.sum, np.mean, np.std])
```

# Filtrado de datos

```python
doble_agrupacion["Edad (años)"].filter(lambda x: x.sum() > 2100)
```

# Transformación de variables

```python
zscore = lambda x : (x - x.mean())/x.std()
```

```python
z_group = doble_agrupacion.transform(zscore)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(z_group["Edad (años)"])
```

```python
fill_na_mean = lambda x : x.fillna(x.mean())
```

```python
doble_agrupacion.transform(fill_na_mean)
```
